```python
import math
import pickle
import numpy as np
import pandas as pd
import xarray as xr
```

```python
from EOF_def import EOF_def
from EOF_def import project_onto_eofs
```

### Inputs first.

```python
#load input data to extrapolate testing ... file location was copied, may need to be changed
infile = open("../processed_vertical_data/vertanom_u.p","rb",)
u_input = pickle.load(infile)  ##vertical U cross section
infile.close()

infile = open("../processed_vertical_data/vertanom_ehf.p","rb",)
ehf_input = pickle.load(infile)  ##ZMehf vertical cross section along longitudes
infile.close()

infile = open("../processed_vertical_data/vertanom_gph.p","rb",)
gph_input = pickle.load(infile)  ##ZMehf vertical cross section along longitudes
infile.close()

##flatten and isolate testing
flat_u_14 = u_input[52:, 19:168, :, :].reshape((10 * 149, 37, 36))
print(flat_u_14.shape)

flat_EHF_14 = ehf_input[52:, 19:168, :, :].reshape((10 * 149, 37, 180))
print(flat_EHF_14.shape)

flat_GPH_14= gph_input[52:, 19:168, :, :].reshape((10 * 149, 37, 180))
print(flat_GPH_14.shape)
```

```python
#load input solvers
infile = open("./solvers/Usolver_14_train.p","rb",)
Usolver = pickle.load(infile)  
infile.close()

infile = open("./solvers/EHFsolver_14_train.p","rb",)
EHFsolver = pickle.load(infile)  ##ZMehf vertical cross section along longitudes
infile.close()

infile = open("./solvers/GPHsolver_14_train.p","rb",)
GPHsolver = pickle.load(infile)  ##ZMehf vertical cross section along longitudes
infile.close()

print("Solvers loaded ...")

## U
U_EOF_nw2d, U_eigenv, U_PC_train = EOF_def(Usolver, 10)
## EHF
EHF_EOF_nw2d, EHF_eigenv, EHF_PC_train = EOF_def(EHFsolver, 45)
## GPH
GPH_EOF_nw2d, GPH_eigenv, GPH_PC_train = EOF_def(GPHsolver, 10)
```

```python
##project EOF solvers from training data onto testing data
U_PC_test = project_onto_eofs(flat_u_14, Usolver, U_EOF_nw2d, U_eigenv)
EHF_PC_test = project_onto_eofs(flat_EHF_14, EHFsolver, EHF_EOF_nw2d, EHF_eigenv)
GPH_PC_test = project_onto_eofs(flat_GPH_14, GPHsolver, GPH_EOF_nw2d, GPH_eigenv)
```

```python
U_PC_test.shape
```

```python
##set array for indexing quickly
#training
def reshape_byidx_eof_train(name,target,end_idx,PCS):
    print(str(name),"Reshape")
    target_reshaped = target.reshape((52,149,PCS))
    target_leadtime = target_reshaped[:,:end_idx,:]
    finished_target = target_leadtime.reshape((52*end_idx),PCS)
    print(finished_target.shape)
    print("######################")
    return finished_target;
#testing
def reshape_byidx_eof_test(name,target,end_idx,PCS):
    print(str(name),"Reshape")
    target_reshaped = target.reshape((10,149,PCS))
    target_leadtime = target_reshaped[:,:end_idx,:]
    finished_target = target_leadtime.reshape((10*end_idx),PCS)
    print(finished_target.shape)
    print("######################")
    return finished_target;
```

```python
##14 days
U_14_train = reshape_byidx_eof_train("U",U_PC_train,149,10)
EHF_14_train = reshape_byidx_eof_train("EHF",EHF_PC_train,149,45)
GPH_14_train = reshape_byidx_eof_train("GPH",GPH_PC_train,149,10)

U_14_test = reshape_byidx_eof_test("U",U_PC_test,149,10)
EHF_14_test = reshape_byidx_eof_test("EHF",EHF_PC_test,149,45)
GPH_14_test = reshape_byidx_eof_test("GPH",GPH_PC_test,149,10)
```

```python
#send out files
pickle.dump(np.concatenate((U_14_train,U_14_test),axis=0), open("U_14.p", 'wb'))
pickle.dump(np.concatenate((EHF_14_train,EHF_14_test),axis=0), open("EHF_14.p", 'wb'))
pickle.dump(np.concatenate((GPH_14_train,GPH_14_test),axis=0), open("GPH_14.p", 'wb'))
```
